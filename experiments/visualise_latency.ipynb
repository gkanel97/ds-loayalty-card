```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('measurements.json', 'r') as f:
    data = json.load(f)
```

```python
measurements_arr = [np.array(x) for x in data.values()]
labels_arr = [x for x in data.keys()]
```

```python
fig, ax = plt.subplots()
ax.violinplot(measurements_arr)
ax.set_xticks(np.arange(1, len(labels_arr) + 1))
ax.set_xticklabels(labels_arr, rotation=10)
ax.set_ylabel('Time (sec)', fontsize=12)
ax.set_xlabel('API Endpoint', fontsize=12)
fig.savefig('latency.png', bbox_inches='tight')
plt.show()
```
